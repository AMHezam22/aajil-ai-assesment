```python
# For each missing Item ID, check if it has a name elsewhere
results = []

for item_id in missing_item_ids:
    # Get all rows for this Item ID
    item_rows = df[df['Item ID'] == item_id]

    # Check if any row has a non-null Item Name
    valid_names = item_rows['Item Name'].dropna()
    valid_names = valid_names[valid_names.str.strip() != ''] if len(valid_names) > 0 else valid_names

    if len(valid_names) > 0:
        # Has name in other rows
        results.append({
            'Item ID': item_id,
            'Status': 'Has name in other rows',
            'Name': valid_names.iloc[0],
            'Missing rows': len(item_rows[item_rows['Item Name'].isna()])
        })
    else:
        # No name found anywhere
        results.append({
            'Item ID': item_id,
            'Status': 'No found',
            'Name': None,
            'Total rows': len(item_rows)
        })

import pandas as pd
results_df = pd.DataFrame(results)
print("\nAnalysis Results:")
print(results_df)
```

```python
# Create a lookup table of Item ID -> Item Name (using first non-null name)
name_lookup = df.groupby('Item ID')['Item Name'].apply(lambda x: x.dropna().iloc[0] if len(x.dropna()) > 0 else None)

# Find Item IDs with missing names in some rows
missing_in_some_rows = df[df['Item Name'].isna()]['Item ID'].unique()

# Check which ones have names available
can_be_filled = []
cannot_be_filled = []

for item_id in missing_in_some_rows:
    if pd.notna(name_lookup[item_id]):
        can_be_filled.append((item_id, name_lookup[item_id]))
    else:
        cannot_be_filled.append(item_id)

print("Can be filled (Item ID, Available Name):")
print(can_be_filled)
print("\nCannot be filled (no name found anywhere):")
print(cannot_be_filled)
print(len(cannot_be_filled))
```

```python
# for those has no names i'll give them just ("Unknown 1,2,3, ...etc") i would use the
# first 7 chars i thought if they are related to factory or something but it shown not
# so i'll just tranlsate it and keep the other unknown, i though of using google translate api
# would be easier and but i'll use llms hardcoded to make sure of the resault and faster rather than using
# llm api

import pandas as pd

# Create temporary columns without modifying original df
item_groups = df['Item ID'].astype(str).str[:7]
name_missing = df['Item Name'].isna() | (df['Item Name'] == '') | (df['Item Name'].str.strip() == '')

# Group analysis
group_analysis = []

for group in item_groups.unique():
    # Get data for this group using boolean indexing
    group_mask = item_groups == group
    group_data = df[group_mask]
    group_missing = name_missing[group_mask]

    # Get unique names for this group (excluding missing ones)
    valid_names = group_data[~group_missing]['Item Name'].unique()
    valid_names = [name for name in valid_names if pd.notna(name) and str(name).strip() != '']

    # Count missing names
    missing_count = group_missing.sum()

    # Total rows in group
    total_rows = len(group_data)

    group_analysis.append({
        'Item_Group': group,
        'Names_Available': valid_names,
        'Number_of_Names': len(valid_names),
        'Missing_Names_Count': missing_count,
        'Total_Rows': total_rows,
        'Percentage_Missing': round((missing_count / total_rows) * 100, 2) if total_rows > 0 else 0
    })

# Convert to DataFrame for better display
results_df = pd.DataFrame(group_analysis)

# Sort by Item_Group
results_df = results_df.sort_values('Item_Group')

print("Group Analysis (First 7 digits of Item ID):")
print("=" * 80)
for _, row in results_df.iterrows():
    print(f"Item Group: {row['Item_Group']}")
    print(f"  Names Available: {row['Names_Available']}")
    print(f"  Number of Different Names: {row['Number_of_Names']}")
    print(f"  Missing Names Count: {row['Missing_Names_Count']}")
    print(f"  Total Rows: {row['Total_Rows']}")
    print(f"  Percentage Missing: {row['Percentage_Missing']}%")
```

```python
import pandas as pd
from groq import Groq
from google.colab import userdata
from tqdm import tqdm
import time

# I asked LLM to get the categories and the translation as a dict
# to keep categories same every item. not pass to the prompt everytime

categories = {
    "H-Beams and Steel Sections": "القطاعات الحديدية على شكل H والقطاعات الفولاذية",
    "Steel Sheets and Plates": "ألواح وصفائح فولاذية",
    "Tubes and Pipes": "أنابيب ومواسير",
    "Wire Mesh and Hardware": "شبك سلكي ومواد تثبيت",
    "HDG Items": "مواد مجلفنة بالغمس الساخن",
    "Concrete Types": "أنواع الخرسانة",
    "Window and Door Hardware": "مستلزمات النوافذ والأبواب",
    "Reinforcement Steel": "حديد تسليح",
    "Electrical Items": "مواد كهربائية",
    "Building Components": "مكونات البناء",
    "Insulation Materials": "مواد عزل",
    "Cables and Wiring": "كابلات وأسلاك",
    "Aluminum Profiles": "قطاعات ألمنيوم",
    "MEP Items": "مواد ميكانيكية وكهربائية وصحية",
    "Plumbing Components": "مكونات السباكة",
    "Industrial Materials": "مواد صناعية"
}


# Get API key from Colab userdata
API_KEY = userdata.get('API_KEY')

# Initialize Groq client
client = Groq(api_key=API_KEY)

def translate_item(item_name):
    """Send item to Groq API for translation and categorization"""

    prompt = f"""Analyze this building material item: {item_name}

Provide ONLY this format and the whole name as given:

Name_en: [English name]
Name_ar: [Arabic name]
Category: [H-Beams and Steel Sections, Steel Sheets and Plates, Tubes and Pipes, Wire Mesh and Hardware, HDG Items, Concrete Types, Window and Door Hardware, Reinforcement Steel, Electrical Items, Building Components, Insulation Materials, Cables and Wiring, Aluminum Profiles, MEP Items, Plumbing Components, or Industrial Materials]"""

    try:
        chat_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            model="openai/gpt-oss-120b",
            temperature=0,  # to keep the ai determinstic, each time same category
            stream=False    # The API waits until the complete response is generated, then returns it all at once
        )

        response=chat_completion.choices[0].message.content
        result = {item.split(':', 1)[0].strip(): item.split(':', 1)[1].strip() for item in response.split('\n')}
        result['category_ar'] = categories.get(result['Category'], 'غير معروف')

        return result
    except Exception as e:
        print(f"Error: {e}")
        return None

def process_dataframe(df):
    """Process entire dataframe"""

    results = []

    print(f"Processing {len(df)} items with Groq API...")

    for item_name in tqdm(df["Item Name"]):

        results.append(translate_item(item_name))

        # Small delay to avoid rate limits
        time.sleep(0.5)

    # Add new columns to dataframe
    results_df = pd.DataFrame(results)
    df["item_name_en"] = results_df["name_en"]
    df["item_name_ar"] = results_df["name_ar"]
    df["category_en"] = results_df["category_en"]
    df["category_ar"] = results_df["category_ar"]

    return df


# Process the dataframe
# It took like 2H which is not functional, so i would use Asyc function, or longer
# prompt but i've limitation for the free acount groq -> i'm facing here a rate limit exceed error so i'll look for other solution also the async funciton
# won't solve the problem
# i would use hugging face but the real models need permession like LLama
# i would use google translation api or any similer apis but i need classification step
# maybe install llama 7b

df_translated = process_dataframe(df)
```
